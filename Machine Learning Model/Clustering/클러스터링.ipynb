## (1) Clustering이란

* ### Clustering  
    - 각 데이터의 유사성을 측정하여, 높은 대상 집단을 분류하고, 군집 간에 상이성을 규명하는 방법  
    - Y가 없을때, 꼭 Classification을 해야 할 때, 군집분석을 사용해야 하겠지만 정확도는 낮음  
    - 고객 segmentation 을 통한 마케팅 활용 방안 / 군집 별 추가 분석 진행  
* ### Clustering의 종류  
    - K-means clustering : 데이터를 사용자가 지정한 k개의 군집으로 나눔  
    - Hierarchical clustering : 나무 모양의 계층 구조를 형성해가는 방법  
    - DBSCAN : k를 설정할 필요없이 군집화를 할 수 있는 방법

## (2) K-means clustering

* ### K-means clustering  
    - 각 군집에 할당된 포인트들의 평균 좌표를 이용해 중심점을 반복적으로 업데이트  
    - Step1 - 각 데이터 포인트 i에 대해 가장 가까운 중심점을 찾고, 그 중심점에 해당하는 군집 할당  
    - Step2 - 할당된 군집을 기반으로 새로운 중심 계산 , 중심점은 군집 내부 점들 좌표의 평균(mean)으로 함  
    - Step3 - 각 클러스터의 할당이 바뀌지 않을 때 까지, 반복
 
    - 초기 point는 random하게 찍는다 -> 그림(b)    
    - (c)에서 할당된 군집에서 새로운 중심 계산한 점 -> 그림(d)
    - 그림(d)에서 할당된 점을 기준으로 새롭게 군집 할당 -> 그림(e)

* ### K-means clustering 단점  
    - 초기 중심 값에 민감한 반응을 보임  
    - 노이즈와 아웃라이어에 민감
    - 군집의 개수 K를 설정하는 것에 어려움  
  
    - 거리를 기반으로 하다 보니 그림과 같은 형태의 데이터일때 k-mean 클러스터링이 잘 안됨  

* ### K-medoids clustering  
    - K-means clustering의 변형으로, 군집의 무게 중심을 구하기 위해 데이터 평균 대신 중간점을 사용(K-means보다 이상치에 강건함)  
    - 아래 그림의 결과 K-medoids의 중앙점이 더 명확 -> 더 좋은 군집을 형성하기에 좋은 가능성 가짐  
    - 그림(a)의 경우 이상치 때문에 중심이 이상한 곳에 생김!
    

## (3) Hierarchical clustering

* ### Hierarchical clustering  
    - 개체들의 가까운 집단부터 순차적/계층적으로 차근차근 묶어 나가는 방식  
    - 모든 개체들 사이의 거리에 대한 유사도 행렬 계산
    - 거리가 인접한 관측치 끼리 cluster 형성  
    - 유사도 행렬 update  
    - 유사한 개체들이 결합되는 dendogram을 통해 시각화 가능  
    - 사전에 군집의 개수를 정하지 않아도 수행가능!  


## (4) DBSCAN clustering

* ### DBSCAN clustering  
    - eps-neighbor와 MinPts를 사용하여 군집을 구성  
        - Eps-neighbor : 학 데이터를 중심으로 입실론 거리 이내의 데이터들을 한 군집으로 구성  
        - MinPts : 한 군집은 MinPts보다 많거나 같은 수의 데이터로 구성  
        => 만약 MinPts보다 적은 수의 데이터가 eps-neighbor를 형성하면 노이스(noise)로 취급  
    - DBSCAN 알고리즘 순서  
        - 아래 그림에서 보라색 점은 다른 군집에 속해있지 않기 떄문에 Outlier  
        - 최소 거리 입실론 이내의 데이터들이 점진적으로 한 군집으로 합쳐지며, 다양한 군집을 형성  
        (군집이 안된 데이터는 노이즈로 취급) => outlier를 detection하는 분야에도 사용 가능
      
    - DBSCAN 파라미터 :MinPts, eps
        - DBSCAN은 군집 분석을 적용하고자 하는 데이터에 대한 이해도가 충분할 때, 파라미터 설정이 쉬움
        - MinPts  
            - minPts = 변수 + 1  
            - minPts는 3이상으로 설정
        - Eps  
            - 너무 작은 경우, 상당 수의 데이터가 노이즈로 구분  
            - 너무 큰 경우, 군집 수가 하나가 될 가능성이 있음
            - 일반적으로 K-nearest neighbor graph의 distance를 그래프로 나타낸 후, 거리가 급격하게 증가하는 지점을 eps로 지정
