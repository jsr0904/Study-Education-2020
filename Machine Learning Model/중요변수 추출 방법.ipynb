## (1) 중요변수 추출 방법

* #### <u>선형회귀</u>
    - 모델의 성능 측정보다 변수의 활용 인자가 무엇이냐 초점을 둘 때 사용  
    - 변수의 유의성은 p-value를 통해 파악 가능
    - 해석법 : '다른 변수가 고정되어 있고 TV광고가 1단위 증가할 때, 매출이 0.046단위 증가'  
      
    
* #### <u>Decision Tree</u>  
    - **독립 변수의 조건**에 따라 **종속변수를 분리**  
    - 해석에 매우 용이!  
    
* #### 복잡한 모델은 해석이 쉽지 않음  
    - Baggging, RandomForest, Gradient Boosting, NN등은 <u>모형에 대한 해석</u>과 <u>prediction에 대한 해석</u>이 **어려움**  
    - Accuracy가 낮고 설명하기 쉬운 모델(Linear Regression, DT) => 주요 인자를 찾고싶을때!  
      **vs**  
      Accuracy가 높고 설명하기 어려운 모델(Ensemnble, NN) => 성능을 높히고 싶을때

* #### Xgboost의 feature importance  
    - 앙상블 모델들은 중요 feature 를 추출할 수 있는 알고리즘이 내장  
    - Weight : 별수 별 데이터를 분리하는데 쓰인 횟수  
    - Cover : 해당 변수로 분리된 데이터의 수로의 가중치  
    - Gain : Feature를 사용 했을때 줄어드는 평균적인 training loss  
    - Xgboost의 feature importance의 문제점 -> 변수 중요도 측정 기준별로 중요변수가 상이!
    
    - 측정 기준을 Weight로 했을 시 Age,Hours per week등이 중요 feature라는 것을 알 수 있음

* #### feature importance의 좋고 나쁨의 기준  
    - Consistency : 즉정 feature가 영향을 많이 가도록 모델을 수정하였다면, 중요도 측정시 해당 feature의 중요도가 줄지 않아야 함!  
      => 같은 데이터 분포 내에서 약간 모델이 바뀌었다고 해서 변수의 중요도가 바뀐다면 "기준"이 잘못 된것  
      => <u>그래서 Consistency가 있어야 한다</u>  
    - Consistency 가 없다면 두 모델의 feature에 대한 비교가 힘들고, feature importance가 높다고 해서 중요하다고 볼 수 없음
      
    - 변수중요도를 보면 각 모델별로 상이한 결과가 도출(inconsistency) => cough변수는 model A보다 B에서 더 중요한 역할을 함

## (2) Shap value

* #### Shap value  
    - The average marginal contribution of a feature value over all possible conditions  
       ( 샘플링 된 다른 feature들 값에 따라 특정 feature의 marginal contribution이 달라져서, 여러번 수행하여 weighted average를 낸다)  
    - Shap value는 consistency한 특성이 존재  
    - 특정 관측치의 j번째 feature로 인한 실제 예측치와 평균예측치의 평균적인 차이  
    - 해석시 주의점 : 어떤 X 를 올리면 Y가 몇 증가 한다 는 해석은 금지 , <u>인과관계 해석이 아닌 현상자체를 해석해야 한다</u>
